```python
# PART 1: Netflix Movies and TV Shows 
```

```python
# 1. Load the dataset
df= pd.read_csv("netflix_titles.csv")
```

```python
df
```

```python
# 2. Display the first 5 
print("First 5 rows:\n", df.head())
```

```python
#2. Display the last 5 rows
print("\nLast 5 rows:\n", df.tail())
```

```python
# 3. How many rows and columns are there in the dataset?
print("\nShape of the dataset:", df.shape)  # (rows, columns)
```

```python
# 4. Number of unique countries where content is produced
unique_countries = df['country'].dropna().str.split(', ')
unique_countries = unique_countries.explode().nunique()
print("\nNumber of unique countries:", unique_countries)
```

```python
# 5. Display all records where the type is 'TV Show'
tv_shows = df[df['type'] == 'TV Show']
print("\nTV Shows:\n", tv_shows)
```

```python
# 6. Convert the `date_added` column to datetime and extract year & month
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
# 7. How many movies were released in 2019?
movies_2019 = df[(df['type'] == 'Movie') & (df['release_year'] == 2019)]
print("\nNumber of movies released in 2019:", len(movies_2019))
```

```python
# 8. Group by rating and show the count of each category
rating_counts = df['rating'].value_counts()
print("\nRating counts:\n", rating_counts)
```

```python
# PART 2: World Population Dataset
```

```python
# Load dataset
df = pd.read_csv('world_population.csv')
```

```python
df
```

```python
# 1. Display the top 10 rows
print(df.head(10))
```

```python
# 2. Check for null values and handle them
print(df.isnull().sum())
df.dropna(inplace=True)
```

```python
# 3. Population of top 5 most populous countries
top5 = df.nlargest(5, '2022 Population')[['Country/Territory', '2022 Population']]
print(top5)
```

```python
# 4. Add GDP per capita column
df['GDP_per_capita'] = df['GDP'] / df['Population']
print("\nWith GDP per capita:\n", df[['Country', 'GDP_per_capita']].head())
```

```python
# 5. Filter countries with population > 100M and area < 500,000 sq km
mask = (df['2022 Population'] > 1e8) & (df['Area (km²)'] < 5e5)
filtered = df[mask][['Country/Territory', '2022 Population', 'Area (km²)']]
print(filtered)
```

```python
# 6. Sort by density in descending order
sorted_density = df.sort_values(by='Density (per km²)', ascending=False)
print(sorted_density[['Country/Territory', 'Density (per km²)']].head(10))
```

```python
# PART 3: Indian Premier League (IPL) Dataset
```

```python
# 1.Load dataset
matches = pd.read_csv('matches.csv')
deliveries=pd.read_csv('deliveries.csv')
```

```python
# 1. Display matches data
print(matches.head())
```

```python
# 2. Most successful team (most wins)
most_successful = matches['winner'].value_counts().idxmax()
print("\nMost successful team:", most_successful)
```

```python
# 3. Total matches played each season
season_match_counts = matches.groupby('season')['id'].count()
print("\nMatches per season:\n", season_match_counts)
```

```python
# 4. Top 5 players with most 'Player of the Match' awards
top_players = matches['player_of_match'].value_counts().head(5)
print("\nTop 5 players:\n", top_players)
```

```python
# 5. Merge matches and deliveries on match_id
merged = pd.merge(deliveries, matches, left_on='match_id', right_on='id')
print("\nMerged dataset:\n", merged.head())
```

```python
# PART 4: Students Performance
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv('StudentsPerformance.csv')
```

```python
# 1. Display average marks in math, reading, writing
avg_math = df['math score'].mean()
avg_read = df['reading score'].mean()
avg_write = df['writing score'].mean()
print("Average Math:", avg_math)
print("Average Reading:", avg_read)
print("Average Writing:", avg_write)
```

```python
# 2. Who performed better in writing – males or females?
grouped = df.groupby('gender')['writing score'].mean()
print("\nAverage Writing Score by Gender:\n", grouped)
```

```python
# 3. Add a new column for total score and percentage
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df['percentage'] = df['total_score'] / 3
print("\nSample Scores:\n", df[['total_score', 'percentage']].head())
```

```python
# 4. Plot bar chart showing average scores grouped by parental level of education
avg_by_parent = df.groupby('parental level of education')[['math score','reading score','writing score']].mean()
avg_by_parent.plot(kind='bar', figsize=(10,6))
plt.title("Avg Scores by Parental Level of Education")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5. Find students scoring above 90% in all subjects (i.e., >90 in each column)
top_students = df[(df['math score'] > 90) &
                  (df['reading score'] > 90) &
                  (df['writing score'] > 90)]
print("\nStudents scoring above 90 in all subjects:\n", top_students)
```
