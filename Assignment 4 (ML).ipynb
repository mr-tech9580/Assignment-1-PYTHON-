```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_customer_churn.csv') 
df.head()
```

```python
sns.countplot(x='churned', data=df)
plt.title("Churn Distribution")
```

```python
sns.countplot(x='subscription_type', hue='churned', data=df)
plt.title("Churn by Subscription Type")
```

```python
sns.histplot(data=df, x='watch_hours', kde=True)
```

```python
numerical_features = ['age', 'watch_hours', 'last_login_days', 'monthly_fee',
                      'number_of_profiles', 'avg_watch_time_per_day', 'churned']

plt.figure(figsize=(10,6))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```

```python
#TASK 1: DATA OVERVIEW
print("🔹 First 5 Rows:")
print(df.head())
print("\n🔹 Missing/Null Values:")
print(df.isnull().sum())
categorical_cols = ['gender', 'subscription_type', 'region']

print("\n🔹 Unique Values in Key Categorical Columns:")
for col in categorical_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
#Task 2: Univariate Analysis 
sns.set(style="whitegrid")
numerical_cols = ['age', 'watch_hours', 'monthly_fee', 'churned']

for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=col, kde=True, bins=30, color="skyblue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
categorical_cols = ['subscription_type', 'gender', 'region',
                    'device', 'payment_method', 'favorite_genre']

for col in categorical_cols:
    if col in df.columns:
        plt.figure(figsize=(6, 4))
        sns.countplot(data=df, x=col, palette="Set2", order=df[col].value_counts().index)
        plt.title(f"Count Plot of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print(f"⚠️ Column '{col}' not found in dataset.")
```

```python
#Task 3: Bivariate Analysis 

sns.set(style="whitegrid")
cols_to_compare = ['subscription_type', 'region', 'device']
for col in cols_to_compare:
    plt.figure(figsize=(8, 5))
    sns.barplot(x=col, y='watch_hours', data=df, estimator='mean', palette='Blues')
    plt.title(f'Average Watch Hours by {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    plt.figure(figsize=(8, 5))
    sns.barplot(x=col, y='monthly_fee', data=df, estimator='mean', palette='Greens')
    plt.title(f'Average Monthly Fee by {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
plt.figure(figsize=(8, 5))
sns.barplot(x='favorite_genre', y='avg_watch_time_per_day', data=df, estimator='mean', palette='Purples')
plt.title('Average Watch Time per Day by Favorite Genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
churn_cols = ['gender', 'region', 'subscription_type', 'payment_method']

for col in churn_cols:
    plt.figure(figsize=(8, 5))
    sns.barplot(x=col, y='churned', data=df, estimator='mean', palette='Oranges')
    plt.title(f'Churn Rate by {col}')
    plt.ylabel('Churn Rate (Proportion)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Task 4: Correlation Analysis 
numerical_cols = ['age', 'watch_hours', 'last_login_days', 'monthly_fee',
                  'number_of_profiles', 'avg_watch_time_per_day', 'churned']
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("🔍 Correlation Heatmap of Numerical Variables")
plt.tight_layout()
plt.show()
```

Insight 1: Watch Hours vs Churn
Insight: Customers with lower watch_hours tend to have higher churn rates.
Recommendation: Increase engagement through personalized recommendations or watch reminders to reduce churn.
Insight 2: Churn by Subscription Type
Insight: Premium users churn less compared to Basic or Standard users.
Recommendation: Highlight the value of Premium plans in marketing and offer upgrade incentives to reduce churn.
Insight 3: Region vs Watch Hours
Insight: Users from North America have the highest average watch hours, while users from Asia have the lowest.
Recommendation: Tailor content and marketing strategies region-wise. For low-engagement regions, promote region-specific content.
Insight 4: Device Preference and Churn
Insight: Customers using TVs and laptops churn less compared to those using mobile devices.
Recommendation: Enhance mobile user experience and introduce mobile-only plans or exclusive features.
Insight 5: Payment Method vs Churn
Insight: Users paying via credit card have the lowest churn, while UPI and wallet users churn more.
Recommendation: Encourage auto-renewal via credit cards or incentivize long-term subscriptions.
Insight 6: Favorite Genre vs Engagement
Insight: Users whose favorite genres are Thriller and Drama have higher average avg_watch_time_per_day.
Recommendation: Produce or recommend more content in these genres to maintain high engagement.
Insight 7: Gender and Churn
Insight: There is no significant difference in churn rate between male and female users.
Recommendation: Focus retention strategies on behavioral attributes (watch time, login frequency) rather than demographics.
